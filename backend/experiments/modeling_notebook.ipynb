# Lagos Traffic Modeling Notebook
This notebook demonstrates baseline model checks for congestion prediction.

```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error

# Example synthetic baseline evaluation
n = 500
y_true = np.clip(np.random.normal(0.45, 0.18, n), 0, 1)
baseline_last = np.roll(y_true, 1)
baseline_last[0] = y_true[0]
rmse = np.sqrt(mean_squared_error(y_true, baseline_last))
pd.DataFrame({'metric': ['baseline_last_rmse'], 'value': [rmse]})
```
